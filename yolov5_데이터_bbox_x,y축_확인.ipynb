인 경우, x와 y가 바운딩 박스 좌상단인지 센터인지 알아보자!

```python
import numpy as np
import cv2
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = '/content/drive/MyDrive/data/KakaoTalk_20211118_211848190.jpg'

image = cv2.imread(path)
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.figure(figsize=(15,15))
plt.imshow(gray, cmap='gray')
gray_image = np.array(gray)
```

```python
gray_image.shape
```

```python
gray_image
```

# 조심할 부분은 이미지는 원래 (W, H, C)이지만 넘파이에선 (축, 행, 렬)이다.

```python
gray_image[1747, 1989] = 255
```

```python
# 픽셀 60개 칠하기

# gray_image[1747-30:1747+30, 1989-30:1989+30] = 225 # 얘는 잘못된 것이다. 이미지 차원과 넘파이 차원은 분명 다르다!
gray_image[1989-30:1989+30, 1747-30:1747+30] = 225
```

```python
plt.figure(figsize=(15,15))
plt.imshow(gray_image,'gray')
plt.show()
```

```python
im = Image.fromarray(gray_image)
plt.imsave(path[:-4]+'_1.jpg', im, cmap='gray')
```
