```python
import ccxt
import requests
import datetime

import FinanceDataReader as fdr
# https://financedata.github.io/posts/finance-data-reader-users-guide.html
# 주가 데이터 가져오기

from pytrends.request import TrendReq
# https://pypi.org/project/pytrends/
# Unofficial API for Google Trends

import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
from pandas_datareader import data as pdr
from datetime import datetime

# body = '''{"startDate":"2021-07-01", "endDate":"2021-08-05",
# timeUnit":"date","keywordGroups":[{"groupName":"cryptocurrency","keywords":["ethereum","bitcoin"]}]}'''

keyword = 'ethereum'
period = 'today 1-m'

trend_obj = TrendReq()
trend_obj.build_payload(kw_list=[keyword], timeframe=period)
trend_df_1 = trend_obj.interest_over_time()

trend_df_1
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(14,5))
trend_df_1['ethereum'].plot()
plt.title("ethereum", size=15)
plt.legend(labels=['ethereum'], loc='upper left')
```

```python
keyword = 'bitcoin' 

trend_obj.build_payload(kw_list=[keyword], timeframe=period)
trend_df_2 = trend_obj.interest_over_time()

trend_df_2
```

```python
plt.subplot(211)
# plt.figure(figsize=(14,5))
trend_df_2['bitcoin'].plot()
plt.title("bitcoin", size=15)
plt.legend(labels=['bitcoin'], loc='upper left')

plt.subplot(212)
# plt.figure(figsize=(14,5))
trend_df_1['ethereum'].plot()
plt.title("ethereum", size=15)
plt.legend(labels=['ethereum'], loc='upper left')

plt.show()
```
