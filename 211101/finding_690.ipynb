```python
def matching_number(value):
  value = str(value)
  for folder in folder_number:
    if value == folder:
      os.chdir(f'/content/drive/MyDrive/data/{folder}')
      count = len(os.listdir())
      return count

df_folder.file_count = df_folder.number.apply(lambda x: matching_number(x))
df_folder
```

```python
merge_df = pd.merge(df_folder, df_total_list, on='number', how='left')
merge_df
```
